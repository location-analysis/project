```python
############ 한글 폰트 지정하는 부분 #################
import matplotlib
from matplotlib import font_manager, rc
import platform
import pandas as pd
import sweetviz as sv

if platform.system() == 'Windows':
# 윈도우인 경우
    font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:    
# Mac 인 경우
    rc('font', family='AppleGothic')
    
matplotlib.rcParams['axes.unicode_minus'] = False   
#그래프에서 마이너스 기호가 표시되도록 하는 설정입니다.
```

```python
# 레포트 생성 --> html 파일로 저장하기  
## 한글이 깨지시는 분은 제일 아래 코드로 실행해주세요
def platform(col, name):
    try:
        report = col.profile_report()
        report.to_file(f'EDA/{name}.html')
    except:
        # pandas_profiling 오류 시 사용
        advert_report = sv.analyze(col)
        #display the report
        advert_report.show_html(f'EDA/{name}.html')
```

```python
platform(gagu_1,'2021_1인가구비율')
platform(ap,'연령별인구현황_연간')
platform(bus, '대구광역시_시내버스 정류소 위치정보')
platform(parking, '대구광역시_주차장정보')
platform(subway, '대구도시철도공사_역별무료자전거운영현황')
platform(school, '대구시 자치구별 대학교 학령인구 통계')
platform(dong, '대구시 행정구역(동별)')
```
