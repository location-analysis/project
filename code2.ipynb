### 역별 무료 자전거 수

역 주소 데이터
https://www.gimi9.com/dataset/www-data-go-kr-data-filedata-15041102/resource/ecb4cc23-a10c-4b17-9e71-9833c35c838a?inner_span=True 

```python
sub_juso = pd.read_csv('data/81. 대구 주소데이터.csv', encoding='cp949')
sub_juso.head()
```

```python
sub_juso.isna().sum()
```

```python
subway['행정구역'] = '대구'
```

```python
## subway '역명' 공백제거 & 지하철 행정구역 추가
for i in range(0,len(subway)):
    subway['역명'][i] = subway['역명'][i].replace(" ", "")
    
    for j in range(0,len(sub_juso)):
        if sub_juso['역명'][j] in subway['역명'][i]:
            subway['행정구역'][i] = sub_juso['지번주소'][j].split(' ')[1]
    
    if subway['행정구역'][i] == '대구':
        print(subway['역명'][i])
```

```python
subway.head()
```

```python
## 역 데이터를 활용한 자전거 수 및 보관대 수 구하기
sub = subway.groupby('행정구역')['자전거(대)','보관대(조)'].sum()
sub['행정구역'] = list(sub.index)
sub.rename(columns={'자전거(대)': '자전거 수', '보관대(조)':'보관대 수'},inplace=True)
```

```python
##  인덱스 초기화
sub.reset_index(drop = True, inplace = True)
sub.reset_index(drop = True, inplace = True)
sub
```

```python
## 자전거 관련 변수 추가
data = pd.merge(data,sub,how='left')
data
```

```python
## data에 '면적' 변수 추가
data['면적'] = 0

for i in range(0,len(data)):
    for j in range(0,len(dong)):
        if data.loc[:,'행정구역'][i] in dong.loc[:,'구 분'][j]:
            data['면적'][i] = dong.loc[:,'면적(㎢)'][j]

data  
```

### 인구밀도 변수 만들기
인구밀도 =총인구(명) / 총면적(㎢)

```python
data['인구밀도'] = data['총 인구'] / data['면적']
data['인구밀도'] = data['인구밀도'].astype(int) ## float or int
#data['인구밀도'] = round(data['인구밀도'])
data
```
